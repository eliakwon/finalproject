```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("housing_ny.csv")
```

```python
house_price = df[df['type'] == 'house']['price']
house_bedroom = df[df['type'] == 'house']['beds']

condo_price = df[df['type'] == 'condo']['price']
condo_bedroom = df[df['type'] == 'condo']['beds']
```

```python
house_bedroom
```

```python
plt.plot(condo_bedroom, condo_price, label = "Condo")
plt.xlabel("Beds")
plt.ylabel("Price (USD)")
plt.title("Average Price (USD) by Beds for Condo")
plt.show()
```

```python
plt.plot(condo_bedroom, condo_price, label = "Condo", color='#54BECA', marker='.', linestyle='--')
plt.plot(house_bedroom, house_price, label = "House", color='#9C257C', marker='o')
plt.xlabel("Beds")
plt.ylabel("Price (USD)")
plt.title("Average Price (USD) by Beds for Condo/House")
plt.legend()
plt.grid(True)
plt.show()
```
